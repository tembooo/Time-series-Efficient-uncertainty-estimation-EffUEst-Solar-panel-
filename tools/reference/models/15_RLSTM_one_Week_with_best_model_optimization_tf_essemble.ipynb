```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import tensorflow as tf
import tensorflow_probability as tfp
from tensorflow.keras import layers, models, optimizers, Input, callbacks
from datetime import timedelta
import gc
import optuna

# --- Force GPU:0 ---
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

# --- Initial Hyperparameters ---
input_window = 168
output_window = 168
batch_size = 32
variance_scale = 1.0
random_seed = 42
lambda_reg = 0.0002
num_models = 5  # Ensemble of 5 models

# --- EDA Modularization ---
def load_and_prepare_data(holidays_file, train_file, val_file, test_file):
    holidays_df = pd.read_excel(holidays_file)
    holidays_df['DateKEY'] = pd.to_datetime(holidays_df['DateKEY'], format='%Y%m%d')
    df_Train = pd.read_excel(train_file)
    df_Val = pd.read_excel(val_file)
    df_Test = pd.read_excel(test_file)
    return holidays_df, df_Train, df_Val, df_Test

def create_enhanced_features(df, holidays_df):
    df['datetime'] = pd.to_datetime(df['datetime'])
    df['is_weekend'] = df['datetime'].dt.weekday.isin([5, 6]).astype(int)
    df['is_holiday'] = df['datetime'].dt.date.isin(holidays_df['DateKEY'].dt.date).astype(int)
    df['is_long_holiday'] = df['is_holiday'].rolling(window=7).sum().ffill()
    if 'air_temperature' in df.columns:
        df['avg_temperature'] = df['air_temperature'].rolling(window=7).mean().ffill()
        df['temp_lag1'] = df['air_temperature'].shift(1).ffill()
    df['hour'] = df['datetime'].dt.hour
    df['weekday'] = df['datetime'].dt.weekday
    df['is_peak_hour'] = df['hour'].isin([8, 9, 17, 18]).astype(int)
    df[selected_features] = df[selected_features].fillna(df[selected_features].mean())
    return df

def analyze_data(df, features):
    print("Data Shape:", df.shape)
    print("Missing Values:\n", df[features].isnull().sum())
    print("Descriptive Stats:\n", df[features].describe())
    plt.figure(figsize=(10, 6))
    df['energy'].plot(title="Energy Over Time")
    plt.xlabel("Datetime")
    plt.ylabel("Energy")
    plt.show()

# --- Selected features ---
selected_features = [
    'air_temperature', 'diffuse_r', 'elspot', 'full_solar', 'global_r',
    'gust_speed', 'relative_humidity', 'sunshine', 'wind_speed',
    'hour', 'weekday', 'is_weekend', 'is_holiday', 'is_long_holiday',
    'avg_temperature', 'temp_lag1', 'is_peak_hour'
]

# --- Load and Analyze Data ---
holidays_df, df_Train, df_Val, df_Test = load_and_prepare_data(
    "3.Holydays .xlsx", "2_weather_data_training_2.xlsx", "1_weather_data_validation_1.xlsx", "0_weather_data_testing_0.xlsx"
)
df_Train = create_enhanced_features(df_Train, holidays_df)
df_Val = create_enhanced_features(df_Val, holidays_df)
df_Test = create_enhanced_features(df_Test, holidays_df)
analyze_data(df_Train, selected_features)

# --- Normalize data ---
scaler_X = StandardScaler()
scaler_y = StandardScaler()
X_train_full = scaler_X.fit_transform(df_Train[selected_features])
X_val_full = scaler_X.transform(df_Val[selected_features])
X_test_full = scaler_X.transform(df_Test[selected_features])
y_train_full = scaler_y.fit_transform(df_Train['energy'].values.reshape(-1, 1))
y_val_full = scaler_y.transform(df_Val['energy'].values.reshape(-1, 1))
y_test_full = scaler_y.transform(df_Test['energy'].values.reshape(-1, 1))

# --- Sliding window ---
def create_sliding_windows(X_full, y_full, input_window, output_window):
    X, y = [], []
    for i in range(len(X_full) - input_window - output_window + 1):
        X.append(X_full[i:i + input_window])
        y.append(y_full[i + input_window:i + input_window + output_window].flatten())
    return np.array(X), np.array(y)

X_train, y_train = create_sliding_windows(X_train_full, y_train_full, input_window, output_window)
X_val, y_val = create_sliding_windows(X_val_full, y_val_full, input_window, output_window)
X_test, y_test = create_sliding_windows(X_test_full, y_test_full, input_window, output_window)
time_test = pd.Series(pd.to_datetime(df_Test['datetime'])[input_window:input_window + len(y_test)])

X_train = X_train.reshape((-1, input_window, len(selected_features)))
X_val = X_val.reshape((-1, input_window, len(selected_features)))
X_test = X_test.reshape((-1, input_window, len(selected_features)))
y_train = y_train.reshape((-1, output_window))
y_val = y_val.reshape((-1, output_window))
y_test = y_test.reshape((-1, output_window))

# --- TensorFlow dataset ---
def create_tf_dataset(X, y, batch_size=32):
    dataset = tf.data.Dataset.from_tensor_slices((X, y))
    return dataset.shuffle(buffer_size=10000).batch(batch_size).prefetch(tf.data.AUTOTUNE)

train_dataset = create_tf_dataset(X_train, y_train, batch_size)
val_dataset = create_tf_dataset(X_val, y_val, batch_size)

# --- Hybrid NLL + MSE loss with peak penalty ---
@tf.function
def hybrid_loss(y_true, y_pred, alpha=0.2):
    mu = y_pred[:, :, 0]
    log_var = y_pred[:, :, 1]
    var = tf.nn.softplus(log_var) + 1e-6
    nll = tf.reduce_mean(0.5 * (tf.math.log(var) + tf.square(y_true - mu) / var))
    mse = tf.reduce_mean(tf.square(y_true - mu))
    peak_mask = y_true > tfp.stats.percentile(y_true, 80.0, axis=-1, keepdims=True)
    mse_peak = tf.where(peak_mask, tf.square(y_true - mu) * 1.5, tf.square(y_true - mu))
    reg = lambda_reg * tf.reduce_mean(log_var ** 2)
    return alpha * nll + (1 - alpha) * tf.reduce_mean(mse_peak) + reg

# --- RLSTM Model ---
def build_rlstm_model(lstm_units, dropout_rate, learning_rate, seed_offset, input_shape, output_window=168):
    tf.random.set_seed(random_seed + seed_offset)
    inputs = Input(shape=input_shape)
    x = layers.LSTM(lstm_units, return_sequences=True, dropout=dropout_rate,
                    kernel_initializer=tf.keras.initializers.GlorotNormal(seed=seed_offset))(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.LSTM(lstm_units // 2, return_sequences=True, dropout=dropout_rate,
                    kernel_initializer=tf.keras.initializers.GlorotNormal(seed=seed_offset+1))(x)
    x = layers.LSTM(lstm_units // 4, return_sequences=False, dropout=dropout_rate,
                    kernel_initializer=tf.keras.initializers.GlorotNormal(seed=seed_offset+2))(x)
    x_residual = layers.Dense(lstm_units // 4, activation='relu')(x)
    x = layers.RepeatVector(output_window)(x + x_residual)
    x = layers.LSTM(lstm_units // 2, return_sequences=True, dropout=dropout_rate,
                    kernel_initializer=tf.keras.initializers.GlorotNormal(seed=seed_offset+3))(x)
    x = layers.LSTM(lstm_units // 4, return_sequences=True, dropout=dropout_rate,
                    kernel_initializer=tf.keras.initializers.GlorotNormal(seed=seed_offset+4))(x)
    x = layers.TimeDistributed(layers.Dense(lstm_units // 4, activation='relu',
                     kernel_initializer=tf.keras.initializers.GlorotNormal(seed=seed_offset+5)))(x)
    outputs = layers.TimeDistributed(layers.Dense(2, kernel_initializer=tf.keras.initializers.GlorotNormal(seed=seed_offset+6),
                          bias_initializer=tf.keras.initializers.Constant(value=[0.0, -3.0])))(x)
    model = models.Model(inputs, outputs)
    model.compile(optimizer=optimizers.Adam(learning_rate),
                  loss=lambda y_true, y_pred: hybrid_loss(y_true, y_pred))
    return model

# --- Optuna Objective Function ---
def objective(trial):
    lstm_units = trial.suggest_int('lstm_units', 512, 2048, step=128)
    dropout_rate = trial.suggest_float('dropout_rate', 0.1, 0.3, step=0.05)
    learning_rate = trial.suggest_float('learning_rate', 1e-5, 1e-3, log=True)
    seed_offset = trial.suggest_int('seed_offset', 10, 50, step=5)

    model = build_rlstm_model(lstm_units, dropout_rate, learning_rate, seed_offset,
                             (input_window, len(selected_features)), output_window)
    model.fit(train_dataset, validation_data=val_dataset, epochs=50,
              callbacks=[callbacks.EarlyStopping(monitor='val_loss', patience=10)],
              verbose=0)
    y_pred_val = model.predict(X_val, batch_size=batch_size)
    mu_pred_val = y_pred_val[:, :, 0]
    y_val_flat = y_val.reshape(-1, output_window)
    mae = mean_absolute_error(y_val_flat, mu_pred_val)
    return mae

# --- Optimize with Optuna ---
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=20)
best_params = study.best_params
print("Best Hyperparameters:", best_params)

# --- Train Ensemble Models ---
ensemble_models = []
for i in range(num_models):
    model = build_rlstm_model(best_params['lstm_units'], best_params['dropout_rate'],
                             best_params['learning_rate'], best_params['seed_offset'] + i * 10,
                             (input_window, len(selected_features)), output_window)
    model.fit(train_dataset, validation_data=val_dataset, epochs=200,
              callbacks=[
                  callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=15, min_lr=1e-6),
                  callbacks.EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True, min_delta=0.001),
                  callbacks.ModelCheckpoint(f'model_{i+1}.h5', monitor='val_loss', save_best_only=True, mode='min')
              ], verbose=2)
    ensemble_models.append(model)
    tf.keras.backend.clear_session()
    gc.collect()

# --- Ensemble Prediction ---
y_pred_ensemble = np.zeros((len(X_test), output_window))
for model in ensemble_models:
    y_pred_mean = model.predict(X_test, batch_size=32)
    y_pred_ensemble += y_pred_mean[:, :, 0]
y_pred_ensemble /= num_models

mu_pred_scaled = y_pred_ensemble
log_var_pred = np.zeros_like(mu_pred_scaled)  # Simplified for ensemble mean
var_pred = np.ones_like(mu_pred_scaled) * 1e-6  # Placeholder variance
std_pred_scaled = np.sqrt(var_pred)
mu_pred = scaler_y.inverse_transform(mu_pred_scaled.reshape(-1, 1)).reshape(mu_pred_scaled.shape)
std_pred = std_pred_scaled * scaler_y.scale_[0]
y_true = scaler_y.inverse_transform(y_test.reshape(-1, 1)).reshape(y_test.shape)

# --- Evaluate Ensemble ---
mu_flat = mu_pred.reshape(-1)
std_flat = std_pred.reshape(-1)
y_true_flat = y_true.reshape(-1)
nll = np.mean(0.5 * (np.log(std_flat**2) + (y_true_flat - mu_flat)**2 / (std_flat**2)) + 0.5 * np.log(2 * np.pi))
mae = mean_absolute_error(y_true_flat, mu_flat)
rmse = np.sqrt(mean_squared_error(y_true_flat, mu_flat))
print(f"\n📊 Ensemble Model Evaluation:")
print(f"MAE:  {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"NLL:  {nll:.4f}")

# --- Plotting function ---
def plot_with_uncertainty(y_true, mu, std, start_idx, end_idx, title, x_range=None):
    plt.figure(figsize=(20, 5))
    if x_range is None:
        x_range = range(start_idx, end_idx)
    max_len = min(len(y_true), len(mu), len(std), len(x_range))
    end_idx = start_idx + max_len
    plt.plot(x_range[start_idx:end_idx], y_true[start_idx:end_idx], label='True Energy', color='blue')
    plt.plot(x_range[start_idx:end_idx], mu[start_idx:end_idx], label='Predicted Mean', color='black')
    plt.fill_between(x_range[start_idx:end_idx],
                     mu[start_idx:end_idx] - 1.96 * std[start_idx:end_idx] * variance_scale,
                     mu[start_idx:end_idx] + 1.96 * std[start_idx:end_idx] * variance_scale,
                     color='orange', alpha=0.3, label='95% CI (Scaled)')
    plt.title(title)
    plt.xlabel("Hour" if x_range is None else "Datetime")
    plt.ylabel("Energy")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_with_uncertainty(y_true_flat, mu_flat, std_flat, 0, 168, "🕐 Ensemble Model - 1 Week", time_test[:168])
plot_with_uncertainty(y_true_flat, mu_flat, std_flat, 0, 24*30*3, "📅 Ensemble Model - 3 Months", time_test[:24*30*3])

# --- Save predictions to Excel ---
df_true = pd.DataFrame(y_test, columns=[f"t+{i+1}" for i in range(output_window)])
df_pred = pd.DataFrame(mu_pred, columns=[f"t+{i+1}" for i in range(output_window)])
df_std = pd.DataFrame(std_pred, columns=[f"t+{i+1}" for i in range(output_window)])
df_true.insert(0, "Week", range(1, len(df_true)+1))
df_pred.insert(0, "Week", range(1, len(df_pred)+1))
df_std.insert(0, "Week", range(1, len(df_std)+1))
with pd.ExcelWriter("ensemble_model_predictions.xlsx") as writer:
    df_true.to_excel(writer, sheet_name="True Energy", index=False)
    df_pred.to_excel(writer, sheet_name="Predicted Energy", index=False)
    df_std.to_excel(writer, sheet_name="Uncertainty (StdDev)", index=False)
print("✅ Excel file 'ensemble_model_predictions.xlsx' with uncertainty saved successfully.")

def compute_coverage(y_true, mu_pred, std_pred, z=1.96):
    lower_bound = mu_pred - z * std_pred
    upper_bound = mu_pred + z * std_pred
    within_bounds = (y_true >= lower_bound) & (y_true <= upper_bound)
    coverage = np.mean(within_bounds) * 100
    return coverage

coverage_percent = compute_coverage(y_true_flat, mu_flat, std_flat)
print(f"📈 Coverage: {coverage_percent:.2f}% of true values are within the 95% prediction interval.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create full time axis for flattened predictions
# time_test originally starts at input_window in df_Test['datetime']
full_time_axis = pd.date_range(start=df_Test['datetime'].iloc[input_window], periods=len(y_true_flat), freq='H')

# Step 2: Plot
def plot_full_test_with_uncertainty(y_true_flat, mu_flat, std_flat, time_axis, title="True vs Predicted Energy (Test Set)", variance_scale=1.0):
    plt.figure(figsize=(22, 6))
    plt.plot(time_axis, y_true_flat, label='True Energy', color='blue', linewidth=1)
    plt.plot(time_axis, mu_flat, label='Predicted Energy', color='green', linewidth=1)

    lower = mu_flat - 1.96 * std_flat * variance_scale
    upper = mu_flat + 1.96 * std_flat * variance_scale
    plt.fill_between(time_axis, lower, upper, color='orange', alpha=0.3, label='95% Confidence Interval')

    plt.title(title)
    plt.xlabel("Datetime")
    plt.ylabel("Energy")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Run the plot
plot_full_test_with_uncertainty(y_true_flat, mu_flat, std_flat, full_time_axis)



import matplotlib.pyplot as plt
import numpy as np

def plot_uncertainty_band_only(std_flat, time_axis, title="Uncertainty Band Width Over Time (±1.96 × Std)", variance_scale=1.0):
    lower = -1.96 * std_flat * variance_scale
    upper =  1.96 * std_flat * variance_scale

    plt.figure(figsize=(22, 6))
    plt.fill_between(time_axis, lower, upper, color='orange', alpha=0.5, label='±1.96 × Std (95% CI)')
    
    plt.title(title)
    plt.xlabel("Datetime")
    plt.ylabel("Uncertainty Range")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Usage
plot_uncertainty_band_only(std_flat, full_time_axis)
```
